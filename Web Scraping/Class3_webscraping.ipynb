#### HTML Elements:
````html
<tagname>Content goes here...</tagname>
The HTML element is everything from the start tag to the end tag:

<h1>My First Heading</h1>
<p>My first paragraph.</p>
````
From: https://www.w3schools.com/html/html_elements.asp
#### HTML Attributes:

HTML elements can be nested (this means that elements can contain other elements).<br  />
All HTML documents consist of nested HTML elements.<br  />
The following example contains four HTML elements (html, body, h1 and p):<br  />
````html
<body id = 'parent'>
    <div id = 'child_1'>I am the parent of 'child_2.'
        <div id = 'child_2'>I am the parent of 'child_3.'
            <div id = 'child_3'> I am the parent of 'child_4.'
                <div id = 'child_4'>I am not a parent. </div>
            </div>
        </div>
    </div>
</body>
````


#### HTML Attributes:
- Use `<a>` tag with `href` attribute to include hiperlinks.
````html
<a href="https://www.google.com">Go to Google</a>
````
- Use the `<img>` tag with `<src>` attribute to insert images.
    
````html
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/c/c3/Python-logo-notext.svg/110px-Python-logo-notext.svg.png">Python logo</img>
````
- Use the tag with `<alt>` attribute to images, in case image is not loaded.
````html
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/c/c3/Python-logo-notext.svg/110px-Python-logo-notext.svg.png" alt="python logo not loaded">Python logo</img>
````
- Use style attribute to change color of a text of a `<p>` element.
````html
<p style="color:blue;">Blue paragraph.</p>
````

- Use tytle attribute to add extra information to a element.
````html
<p style="color:blue;" title="color_paragraph">Blue paragraph.</p>
````

- Use `<h1>` to `<h6>` for headings from highest to lowest importance.
````html
<h1>Heading 1</h1>
<h2>Heading 2</h2>
<h3>Heading 3</h3>
<h4>Heading 4</h4>
<h5>Heading 5</h5>
<h6>Heading 6</h6>
````

# WIKIPEDIA GDP

`````python
site = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'

page_wiki = requests.get(site)
soup_wiki = BeautifulSoup(page_wiki.content, 'html.parser')

wiki_tables = soup_wiki.findAll("table")
print(len(wiki_tables))
`````

`````python
i = 1
for table in wiki_tables:
    print(i, str(table.attrs))
    i += 1
`````

`````python
gdp_tables = soup_wiki.findAll("table", class_="wikitable")
len(gdp_tables)
`````

`````python
gdp_tables = soup_wiki.find_all("table", class_="wikitable sortable")

len(gdp_tables)
`````

`````python
data_frames = []
for table in gdp_tables:
    df = pd.read_html(str(table))[0]
    display(df.head(3))
    data_frames.append(df)
`````
